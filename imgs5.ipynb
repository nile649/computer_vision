```python
import os
import math
import cv2

listing = os.listdir("/home/n/Documents/plz/")
count = 1
for file in listing:
    video = cv2.VideoCapture("/home/n/Documents/plz/" + file)
    print(video.isOpened())
    framerate = video.get(5)
    os.makedirs("/home/n/Documents/plz/" + "video_" + str(int(count)))
    while (video.isOpened()):
        frameId = video.get(1)
        success,image = video.read()
        if( image != None ):
            image=cv2.resize(image,(224,224), interpolation = cv2.INTER_AREA)
        if (success != True):
            break
        if (frameId % math.floor(framerate) == 0):
            filename = "/home/n/Documents/plz/" + str(int(count)) + "/image_" + str(int(frameId / math.floor(framerate))+1) + ".jpg"
            print(filename)
            cv2.imwrite(filename,image)
    video.release()
    print('done')
count+=1
```

```python
import cv2

def frame_capture(path,lol):
    vidcap = cv2.VideoCapture(path)
    success,image = vidcap.read()
    count = 0
    success = True
    while success:
        success,image = vidcap.read()
        if count%10 ==0:
            cv2.imwrite("/home/n/Documents/videos/test/Walk_%d.jpg" % lol, image)     # save frame as JPEG file      
            lol +=1
            print('Read a new frame: ', success)
        
        count += 1
    return lol
```
